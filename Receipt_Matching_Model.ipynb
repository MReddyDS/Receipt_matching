## Receipt Matching Data Science Challenge

#### Objective
Develop and implement a machine learning model to enhance receipt-image-to-transaction matching. The model aims to identify the most relevant matching features between extracted receipt data and existing transaction records.

```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, auc, precision_recall_curve, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from lightgbm import LGBMClassifier
```

#### Process the input dataset or source file  'data_interview_test.csv'
- Import the column names from the first row of source dataset
- Import the data from 2nd row onwards
- process the independent variables from the matching vectors seperated by the **':'**
- Target variable will be created by matching transaction Id from the **‘matched_transaction_id’** (matched transaction id is combination transaction_id:rceipt_Id) with **‘feature_transaction_id’**.
- Traget value is '1' if *transaction_id == feature transaction_id* otherwise '0'

```python
def data_loader(file_name = 'data_interview_test.csv'):
    
    '''
    This function is to read csv file and processing matchong field separated by ":" delimiter and creating target variable
    
    input - file_nale : CSV file name
    
    output : output pandas dataframe with all the features and target variable
    
    '''

    # Read just the first row (column headers) from the CSV file
    column_names = pd.read_csv(file_name, header=None, nrows=1)[0].str.split(":")[0]
    
    # Read the data skipping header line
    df = pd.read_csv(file_name, header=None, skiprows=1)
    df.columns = column_names[:3]+['OriginalText']
    
    #Splitting OriginalText filed to extract features
    new_columns = df.loc[:,'OriginalText'].str.split(':', expand=True)
    new_columns.columns = column_names[3:]
    
    #concate original dataset with created features
    df = pd.concat([df, new_columns], axis=1)
    
    #Target variable creation by comparing matching transaction_id with feature transaction id
    df['Target'] = np.where(df['matched_transaction_id'].str.split(":", n=1).str[0] == df['feature_transaction_id'], 1, 0)
    return df
```

```python
# load data and print top 5 records
data = data_loader(file_name = 'data_interview_test.csv')
data.head()
```

### Modeling Aproches
- Since the data is highly imbalanced we can't build classification model with regular approach. I have tried following approches.

##### 1. Resampling or Undersampling
- When using this approach, we consider undersampling the majority class. However, in this case, it is not recommended because the number of examples in the minority class is very small. If we undersample the majority class, the total number of training examples for the model will be greatly reduced, potentially resulting in a model that lacks generalization ability.

##### 2. Synthetic Examples
- Another technique involves generating synthetic examples using SMOTE (Synthetic Minority Over-sampling Technique). However, this approach also has limitations because of the scarcity of examples in the minority class. The synthesized records might lack variety, potentially leading to overfitting.

##### 3. Class Weights
- This approach has shown better performance compared to the aforementioned methods. Scikit-learn, XGBoost, and LightGBM have built-in class weight parameters that assign higher weightage to the minority class during training.


###### Generally, tree-based models perform well in the case of imbalanced datasets. Therefore, I tried Random Forest, XGBoost, and LightGBM. Interestingly, all three models yielded similar results, and I ultimately chose the LightGBM model for further analysis..

#### Building Light GBM Classifier
- Since the data is imbalanced, using stratified cross-validation will help mitigate overfitting of the majority class.
- LightGBM has parameters 'is_unbalance' and 'scale_pos_weight' that allow passing class weights for binary classification problems. In this case, 'is_unbalance = True' was used.

```python
# Step 1: Define parameter grid for hyperparameter tuning
param_grid = {
    'boosting_type' : ['gbdt', 'dart', 'rf'],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [50, 100, 150, 200, 250, 300],
    'num_leaves' : [10, 20, 30, 31, 40, 50],
    'max_depth': [2, 4, 6, 8, 10],
    'colsample_bytree': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1],
    'subsample': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9],
    'min_child_samples': [1, 5, 10]
}

# Step 2: Initialize LGBMClassifier estimattor
estimator = LGBMClassifier(objective='binary',
                              is_unbalance = True,
                              verbose = -1,
                              random_state=102)



# Step 3: Initalise RandomizedSearchCV Search with 5-fold cross validation and fit model
model_lgb = RandomizedSearchCV(estimator=estimator, 
                     param_distributions=param_grid,
                     cv=StratifiedKFold(n_splits=5), 
                     n_jobs=-1,
                     random_state=103,
                     scoring='precision')
model_lgb.fit(X_train, y_train)

# Step 4: Print best parameters
best_params = model_lgb.best_estimator_
print(best_params)
```

##### ROC vs Precision Recall curves

- ROC curves plot the TPR (True positive rate) against the FPR (False positive rate) at various classification thresholds. In imbalanced datasets, where the majority class heavily outweighs the minority class, the ROC curve can emphasize the true negative rate (TNR) because the majority class contributes more to the overall FPR. As a result, the curve might not adequately capture the model's performance on the minority class.

- In imbalanced datasets, precision-recall curves and F1-score are often more informative. Precision-recall curves focus on the trade-off between precision and recall, which is particularly important in cases where the minority class is of high interest. F1-score balances both precision and recall and provides a single metric that considers both false positives and false negatives.


##### Create ROC Curve and Precision and Recall functions to generate plots

```python
def create_pr_curve(model, X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test):
    # Predict probabilities for the positive class
    y_prob_traiin = model.predict_proba(X_train)[:, 1]
    y_prob = model.predict_proba(X_test)[:, 1]

    # Calculate precision-recall curve
    precision_train, recall_train, thresholds_train = precision_recall_curve(y_train, y_prob_traiin)
    precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

    # Plot precision-recall curve
    plt.figure()
    plt.step(recall_train, precision_train, color='b', alpha=1, where='post', label ='Train - AP={:.2f}'.format(auc(recall_train, precision_train)))
    plt.step(recall, precision, color='orange', alpha=1, where='post', label='Valid - AP={:.2f}'.format(auc(recall, precision)) )

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.show()
    
def create_roc_curve(model, X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test):
    # Predict probabilities for the positive class
    y_prob_train = model.predict_proba(X_train)[:, 1]
    y_prob = model.predict_proba(X_test)[:, 1]

    # Calculate ROC curve
    fpr_t, tpr_t, thresholds_t = roc_curve(y_train, y_prob_train)
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr_t, tpr_t, color='b', lw=2, label='train ROC curve (area = {:.2f})'.format(auc(fpr_t, tpr_t)))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='Valid ROC curve (area = {:.2f})'.format(auc(fpr, tpr)))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

##### Create ROC curve and check the performance of model

```python
create_roc_curve(model=model_lgb)
```

- It is evident that the model is performing well in both training and validation, with 0.97 area under the curve. However, as previously mentioned, the ROC curve is not always a reliable metric when dealing with imbalanced datasets.
- This is due to the fact that the FPR (False Positive Rate) is influenced by the numerous instances of the majority negative class, potentially leading to a distortion in the evaluation of the ROC curve's performance.
